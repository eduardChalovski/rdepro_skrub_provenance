```python


#this dataframe is only for testing, this should work with any df
df1 = pd.DataFrame ({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"]
    
})

df2 = pd.DataFrame ({
    "id": [1, 2, 4],
    "age": [24, 30, 50]
})
```

```python
result = pd.merge(df1, df2, on="id", how="outer")

pd.set_option('display.max_columns', None)

# Show all rows
pd.set_option('display.max_rows', None)

# Show full column width (no truncation)
pd.set_option('display.max_colwidth', None)
display(result)
```

```python
result = pd.merge(df1, df2, on="id", how="left")
display(result)
```

```python
resultPrime = pd.merge(df1, df2, on="id", how="right")
display(resultPrime)
```
